Identificação das ciadades existentes nos datasets

```python
import pandas as pd
import re

def extrair_cidades_unicas():
    # Nomes dos seus arquivos
    arquivo_br = "Campeonato Brasileiro Série A 2020 até 2025.csv"
    arquivo_pl = "Campeonato Premier League 2019 até 2024.csv"
    
    todas_cidades = set() # Usamos um Set (conjunto) para garantir que não haja repetições
    
    arquivos = [
        {"nome": arquivo_br, "liga": "Brasileirão"},
        {"nome": arquivo_pl, "liga": "Premier League"}
    ]
    
    for arquivo in arquivos:
        try:
            print(f"Lendo o arquivo: {arquivo['nome']}")
            df = pd.read_csv(arquivo['nome'])
            
            # Garante que a coluna existe antes de tentar extrair
            if 'Estadio_Local' in df.columns:
                # Extrai o texto que está dentro dos parênteses
                df['Cidade'] = df['Estadio_Local'].str.extract(r'\((.*?)\)')[0]
                
                # Para jogos que por acaso não têm cidade entre parênteses, usa o nome do mandante como fallback
                df['Cidade'] = df['Cidade'].fillna(df['Mandante'])
                
                # Pega as cidades únicas deste arquivo e limpa espaços em branco extras
                cidades_liga = df['Cidade'].dropna().str.strip().unique()
                
                print(f"-> Encontradas {len(cidades_liga)} cidades únicas na {arquivo['liga']}.")
                
                # Adiciona ao conjunto geral
                for cidade in cidades_liga:
                    todas_cidades.add(cidade)
            else:
                print(f"-> AVISO: Coluna 'Estadio_Local' não encontrada no arquivo {arquivo['nome']}.")
                
        except FileNotFoundError:
            print(f"-> ERRO: O arquivo '{arquivo['nome']}' não foi encontrado.")
            
    # Converte o set para uma lista e ordena em ordem alfabética para facilitar a leitura
    lista_final = sorted(list(todas_cidades))
    
    print(f"\nTotal de cidades únicas (juntando as duas ligas): {len(lista_final)}")
    
    # Salva o resultado em um novo CSV simples
    df_resultado = pd.DataFrame(lista_final, columns=['Cidade'])
    df_resultado.to_csv("lista_todas_cidades.csv", index=False)
    
    print("Arquivo 'lista_todas_cidades.csv' criado com sucesso!")

if __name__ == "__main__":
    extrair_cidades_unicas()
```
